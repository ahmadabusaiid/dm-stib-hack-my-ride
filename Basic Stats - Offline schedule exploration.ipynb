```python
##GitLab ref: https://gitlab.com/mrcagney/gtfs_kit/-/blob/master/notebooks/examples.ipynb
import gtfs_kit as gk
import matplotlib.pyplot as plt
import geopandas as gpd
import shapefile
##want to ignore warnings related to shapely being deprecated .. nothing to worry abt 
import shapely
import warnings
from shapely.errors import ShapelyDeprecationWarning
warnings.filterwarnings("ignore", category=ShapelyDeprecationWarning) 
```

```python
#Declare the directory path for the GTFS zip file
path = 'gtfs3Sept.zip' ##make sure you have the zip file in the same directory

#Read the feed with gtfs-kit
feed = (gk.read_feed(path, dist_units='km'))

#Search for errors and warnings in the feed
feed.validate()
```

```python
trip_stats = feed.compute_trip_stats() ##get the trip statistics 
```

```python
trip_stats.T ## view 
```

```python
DATE = '20210903' ##original schedule date 3 sept 2021
route_stats = feed.compute_route_stats(trip_stats,dates=[DATE])
```

```python
route_stats.T
```

```python
##to ease understanding

route_stats = route_stats.merge(feed.routes.filter(["route_id","route_long_name"]))
cols = [
    "route_id",
    "route_short_name",
    "route_long_name",
    "route_type",
    "start_time",
    "end_time",
    "num_trips",
    "max_headway",
    "mean_trip_distance",
    "mean_trip_duration",
    "service_speed",

]
```

```python
##least freq routes 
##small max_headways

(
    route_stats
    .filter(cols)
    .sort_values("max_headway")
    .T
)
```

```python
##short/long distances
(
    route_stats
    .filter(cols)
    .sort_values("mean_trip_distance")
    .T
)
```

```python
##service speed

(
    route_stats
    .filter(cols)
    .sort_values("service_speed")
    .T
)
```
