```python
##GitLab ref: https://gitlab.com/mrcagney/gtfs_kit/-/blob/master/notebooks/examples.ipynb
import gtfs_kit as gk
##want to ignore warnings related to shapely being deprecated .. nothing to worry abt 
import shapely
import warnings
from shapely.errors import ShapelyDeprecationWarning
warnings.filterwarnings("ignore", category=ShapelyDeprecationWarning) 
```

```python
#Declare the directory path for the GTFS zip file
path = 'gtfs3Sept.zip'

#Read the feed with gtfs-kit
feed = (gk.read_feed(path, dist_units='km'))

#Search for errors and warnings in the feed
feed.validate()
```

```python
feed.describe()
```

```python
display(feed.stop_times.head().T)

feed = feed.append_dist_to_stop_times()
feed.stop_times.head().T
```

```python
trip_stats = feed.compute_trip_stats() ##get the trip statistics 
```

```python
trip_stats.T ## view 
```

```python
##schedule includes 
##3,6,7,8,10,13,14,15,16,17 -> weekday
##11,12,18,19 -> weekend 

##pick date interval to compute stats 

week = feed.get_first_week()
dates = [week[0],week[1], week[2],week[3],week[4]]  # get the week days for teh first week
dates
```

```python
route_stats = feed.compute_route_stats(trip_stats,dates=dates)
```

```python
route_stats.T ## shape adds up with the number of lines STIB operates 
```

```python
##to ease understanding

route_stats = route_stats.merge(feed.routes.filter(["route_id","route_long_name"]))
cols = [
    "route_id",
    "route_short_name",
    "route_long_name",
    "route_type",
    "start_time",
    "end_time",
    "num_trips",
    "max_headway",
    "mean_trip_distance",
    "mean_trip_duration",
    "service_speed",

]
```

```python
##least freq routes 
##small max_headways

(
    route_stats
    .filter(cols)
    .sort_values("max_headway")
    .T
)
```

```python
##short/long distances
(
    route_stats
    .filter(cols)
    .sort_values("mean_trip_distance")
    .T
)
```

```python
##service speed

(
    route_stats
    .filter(cols)
    .sort_values("service_speed")
    .T
)
```
